###3. Plotting a graph between male and female participants

```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# importing the required libraries for plotting a count plot 
```

```python
sns.countplot(x='Buyer Gender',data = df)
plt.title('Coomparison Between Male & Female Participants',fontsize = 15) 
plt.xlabel('Buyer Gender') 
plt.ylabel('No of Participants')
plt.yticks()
plt.grid()
```

```python
# We can see that they are fairly identical in number, but female participants apprear to slightly more. Checking the exact count of values, next
```

```python
df["Buyer Gender"].count()
```

```python
participant_count = df["Buyer Gender"].value_counts()
participant_count
```

```python
# From this, it is clear that female buyers are more than male buyers by a value of 104
```

###4. Finding the top 5 cars based on their sales price

```python
top5 = df.sort_values(by=["Sale Price"],ascending= False)
top5.iloc[:5]
```

```python
# The top 5 models and their sale price are as follows
```

```python
# creting a separate dataframe to store required data and then dropping rows after the 5th index
```

```python
top5_df = pd.DataFrame(top5,columns=['Make','Model','Sale Price'])
top5_df.iloc[:5]
```

###5. Finding the top 5 cars based on lowest reselling price

```python
# Followed the same procedure as above, but this time, selected "Resell price" and kept the ascending value as True
```

```python
top5_resell = df.sort_values(by=["Resell Price"])
top5_resell_df = top5_resell.iloc[:5]
top5_resell_df = pd.DataFrame(top5_resell,columns=['Make','Model','Resell Price'])
top5_resell_df.iloc[:5]
```
